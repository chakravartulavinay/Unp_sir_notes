# Introduction to Matplotlib

In this notebook, we will cover the basic syntax of using Matplotlib, explore its different functions, and learn about the various types of graphs we can generate using this powerful data visualization library.

## 1. Basic Syntax of Matplotlib

Matplotlib is a 2D plotting library for Python. It enables us to create a wide range of plots, such as line plots, bar plots, scatter plots, histograms, and more. Let's begin by importing the necessary libraries and creating a simple line plot.

````python
import matplotlib.pyplot as plt

# Synthetic data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10] # y = 2x(Line)

# Create a basic line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
````

Matplotlib is a popular data visualization library in Python that allows you to create various types of 2D plots and graphs. It provides a wide range of functions and options to customize your plots according to your needs. Here's an in-depth explanation of the basic syntax of Matplotlib:

1. Importing the Library:
   The first step is to import the `matplotlib.pyplot` module, which is the main interface for creating plots in Matplotlib. It is usually imported using the alias `plt`, which is a common convention:

   ```python
   import matplotlib.pyplot as plt
   ```

2. Creating a Figure and Axes:
   In Matplotlib, a "figure" is the container that holds all the elements of a plot, and "axes" are the actual plots within the figure. You can create a figure and axes using the `plt.subplots()` function:

   ```python
   fig, ax = plt.subplots()
   ```

3. Plotting Data:
   Once you have the axes, you can use various plot types like `plot()`, `scatter()`, `bar()`, `pie()`, etc., to visualize your data. For example, to create a simple line plot, you can use the `plot()` function:

   ```python
   x = [1, 2, 3, 4, 5]
   y = [2, 4, 6, 8, 10]

   ax.plot(x, y)
   ```

4. Adding Labels and Title:
   You can add labels to the x-axis and y-axis using the `xlabel()` and `ylabel()` functions, respectively. Additionally, you can set a title for the plot using the `title()` function:

   ```python
   ax.set_xlabel('X-axis Label')
   ax.set_ylabel('Y-axis Label')
   ax.set_title('Plot Title')
   ```

5. Customizing Plot Appearance:
   Matplotlib provides many options to customize the appearance of your plot, such as changing line styles, marker styles, colors, grid lines, etc. You can use different arguments in the plot functions to achieve this:

   ```python
   # Example: Customizing the plot appearance
   ax.plot(x, y, linestyle='--', marker='o', color='blue')
   ax.grid(True)  # Show grid lines
   ```

6. Showing the Plot:
   After customizing your plot, you can use the `plt.show()` function to display it. This function should be called only once at the end of your code, as it blocks further execution until the plot window is closed:

   ```python
   plt.show()
   ```

And that's the basic syntax of creating a simple plot using Matplotlib. You can use the same basic steps to create other types of plots and customize them according to your data and visualization requirements. Remember that there are many more advanced features and options available in Matplotlib, but the basics covered above should help you get started with creating basic visualizations.

## 2. Different Functions of Matplotlib

Matplotlib provides a wide range of functions to customize plots. Let's explore some common functions like `plt.xlabel()`, `plt.ylabel()`, `plt.title()`, and `plt.legend()` using bar plots.

````python
# Synthetic data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 50]

# Create a bar plot
plt.bar(categories, values)
# plt.yticks([0,5,10,15,20,25,30,35,40,45,50])
# plt.yticks(range(0,51,5))
# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')

# Show legend
plt.legend(['Values'])
plt.show()
````

Matplotlib provides a wide range of functions that allow you to customize and enhance your plots. These functions help you control various aspects of the plot, such as axis labels, titles, legends, colors, styles, and more. Here's a detailed explanation of some commonly used functions in Matplotlib:

1. `plt.xlabel()`, `plt.ylabel()`, `plt.title()`: These functions are used to set the labels for the x-axis, y-axis, and the title of the plot, respectively. For example:

   ```python
   plt.xlabel('X-axis Label')
   plt.ylabel('Y-axis Label')
   plt.title('Plot Title')
   ```

2. `plt.legend()`: This function is used to add a legend to your plot, which helps identify different elements on the plot. You can pass a list of labels to `plt.legend()` to indicate what each element represents:

   ```python
   plt.plot(x1, y1, label='Line 1')
   plt.plot(x2, y2, label='Line 2')
   plt.legend()
   ```

3. `plt.grid()`: The `plt.grid()` function allows you to display grid lines on the plot, making it easier to read the data:

   ```python
   plt.plot(x, y)
   plt.grid(True)  # Show grid lines
   ```

4. `plt.xticks()`, `plt.yticks()`: These functions allow you to customize the tick positions and labels on the x-axis and y-axis, respectively. You can pass a list of positions and labels to these functions:

   ```python
   plt.xticks([1, 2, 3, 4, 5], ['One', 'Two', 'Three', 'Four', 'Five'])
   plt.yticks([0, 5, 10, 15, 20])
   ```

5. `plt.xlim()`, `plt.ylim()`: These functions let you set the limits of the x-axis and y-axis, respectively:

   ```python
   plt.xlim(0, 10)
   plt.ylim(0, 100)
   ```

6. `plt.plot()` and `plt.scatter()`: These functions are used to create line plots and scatter plots, respectively. The `plot()` function is suitable for creating line plots, while the `scatter()` function is used for scatter plots:

   ```python
   plt.plot(x, y)  # Line plot
   plt.scatter(x, y)  # Scatter plot
   ```

7. `plt.bar()`, `plt.hist()`, `plt.pie()`: These functions are used to create bar plots, histograms, and pie charts, respectively:

   ```python
   plt.bar(categories, values)  # Bar plot
   plt.hist(data, bins=10)  # Histogram
   plt.pie(sizes, labels=labels, autopct='%1.1f%%')  # Pie chart
   ```

8. `plt.savefig()`: This function allows you to save the plot as an image file. You can specify the filename and file format:

   ```python
   plt.plot(x, y)
   plt.savefig('plot.png', dpi=300)  # Save plot as a PNG image with 300 DPI
   ```

9. `plt.clf()`: This function clears the current figure, allowing you to create a new plot without creating a new figure:

   ```python
   plt.plot(x1, y1)
   plt.show()

   plt.clf()  # Clear the current figure

   plt.plot(x2, y2)
   plt.show()
   ```

These are just some of the many functions available in Matplotlib. You can explore the official Matplotlib documentation for a comprehensive list of functions and their usage. Matplotlib provides extensive functionality, making it a powerful tool for creating professional-quality visualizations in Python.

## 3. Different Types of Graphs using Matplotlib

Matplotlib allows us to create various types of graphs. Some common types include line plots, scatter plots, bar plots, pie charts, and histograms. Let's generate examples for each type using synthetic data.

Matplotlib provides various types of plots to visualize different types of data. Let's explore some of the most commonly used plot types along with examples:

1. **Line Plot:**
   A line plot is used to visualize the relationship between two variables with continuous data points. It is created using the `plt.plot()` function.

   ```python
   import matplotlib.pyplot as plt

   # Sample data
   x = [1, 2, 3, 4, 5]
   y = [2, 4, 6, 8, 10]

   # Create a line plot
   plt.plot(x, y)
   plt.xlabel('X-axis')
   plt.ylabel('Y-axis')
   plt.title('Line Plot')
   plt.show()
   ```

2. **Scatter Plot:**
   A scatter plot is used to show the relationship between two variables with discrete data points. Each data point is represented by a marker on the plot. Scatter plots are created using the `plt.scatter()` function.

   ```python
   import matplotlib.pyplot as plt

   # Sample data
   x = [1, 2, 3, 4, 5]
   y = [2, 4, 6, 8, 10]

   # Create a scatter plot
   plt.scatter(x, y)
   plt.xlabel('X-axis')
   plt.ylabel('Y-axis')
   plt.title('Scatter Plot')
   plt.show()
   ```

3. **Bar Plot:**
   Bar plots are used to compare different categories or discrete data. They are created using the `plt.bar()` function.

   ```python
   import matplotlib.pyplot as plt

   # Sample data
   categories = ['Category A', 'Category B', 'Category C', 'Category D']
   values = [25, 40, 30, 50]

   # Create a bar plot
   plt.bar(categories, values)
   plt.xlabel('Categories')
   plt.ylabel('Values')
   plt.title('Bar Plot')
   plt.show()
   ```

4. **Histogram:**
   Histograms are used to visualize the distribution of continuous data. They are created using the `plt.hist()` function.

   ```python
   import matplotlib.pyplot as plt

   # Sample data
   data = [10, 20, 30, 40, 50, 20, 30, 40, 20, 30, 10, 15]

   # Create a histogram
   plt.hist(data, bins=5, edgecolor='black')
   plt.xlabel('Values')
   plt.ylabel('Frequency')
   plt.title('Histogram')
   plt.show()
   ```

5. **Pie Chart:**
   Pie charts are used to represent the proportions of different categories as a part of a whole. They are created using the `plt.pie()` function.

   ```python
   import matplotlib.pyplot as plt

   # Sample data
   labels = ['A', 'B', 'C', 'D']
   sizes = [15, 30, 45, 10]

   # Create a pie chart
   plt.pie(sizes, labels=labels, autopct='%1.1f%%')
   plt.title('Pie Chart')
   plt.show()
   ```

6. **Box Plot:**
   Box plots are used to show the distribution of data and identify outliers. They are created using the `plt.boxplot()` function.

   ```python
   import matplotlib.pyplot as plt

   # Sample data
   data = [10, 20, 25, 30, 35, 40, 50]

   # Create a box plot
   plt.boxplot(data)
   plt.ylabel('Values')
   plt.title('Box Plot')
   plt.show()
   ```

7. **Heatmap:**
   Heatmaps are used to visualize 2D data as a color-coded grid. They are created using the `plt.imshow()` function.

   ```python
   import numpy as np
   import matplotlib.pyplot as plt

   # Sample data
   data = np.random.rand(5, 5)

   # Create a heatmap
   plt.imshow(data, cmap='viridis')
   plt.colorbar()
   plt.title('Heatmap')
   plt.show()
   ```

These are just a few examples of the different types of plots you can create using Matplotlib. The library offers many more plot types and customization options, allowing you to create a wide range of visualizations for your data.

````python
import pandas as pd
import matplotlib.pyplot as plt
path = "" # give path of your csv file
# Load the dataset
df = pd.read_csv(path)
````

````python
# Display the first few rows of the dataset
df.head()
````

````python
df.describe()
````

````python
# Extract data for line plot
months = df['Month']
sales = df['Sales']

# Create a line plot
plt.plot(months, sales, marker='o', linestyle='-', color='b')

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.grid(True)
plt.show()
````

````python
# Extract data for scatter plot
expenses = df['Expenses']
units_sold = df['Units Sold']

# Create a scatter plot
plt.scatter(expenses, units_sold, marker='o', color='g')

# Customize the plot
plt.xlabel('Expenses')
plt.ylabel('Units Sold')
plt.title('Expenses vs. Units Sold')

# Show the plot
plt.grid(True)
plt.show()
````

````python
# Extract data for bar plot
profits = df['Profit']

# Create a bar plot
plt.bar(months, profits)

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Profit')
plt.title('Monthly Profit')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.grid(True)
plt.show()
````

````python
# Extract data for histogram
sales = df['Sales']

# Create a histogram
plt.hist(sales, bins=5, edgecolor='black')

# Customize the plot
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Sales Distribution')

# Show the plot
plt.grid(True)
plt.show()
````

````python
# Extract data for pie chart
categories = ['Sales', 'Expenses', 'Profit']
totals = [df['Sales'].sum(), df['Expenses'].sum(), df['Profit'].sum()]

# Create a pie chart
plt.pie(totals, labels=categories, autopct='%1.1f%%')

# Customize the plot
plt.title('Financial Overview')

# Show the plot
plt.show()
````

````python
# Extract data for box plot
sales_data = [df['Sales'], df['Expenses'], df['Profit'], df['Units Sold']]

# Create a box plot
plt.boxplot(sales_data, labels=['Sales', 'Expenses', 'Profit', 'Units Sold'])

# Customize the plot
plt.ylabel('Values')
plt.title('Financial Metrics')

# Show the plot
plt.grid(True)
plt.show()
````

````python
# Extract data for heatmap
heatmap_data = df[['Sales', 'Expenses', 'Profit', 'Units Sold']].corr()

# Create a heatmap
plt.imshow(heatmap_data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()

# Customize the plot
plt.xticks(range(len(heatmap_data)), heatmap_data.columns, rotation=45)
plt.yticks(range(len(heatmap_data)), heatmap_data.columns)
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
````
